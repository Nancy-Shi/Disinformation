##Visualization of Parameter Approximation Result by DNN##

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
```

```python
df = pd.read_csv('/content/drive/My Drive/Misinformation_Project/neuralepi12.csv')
```

```python
# import libraries
import matplotlib.dates as mdates

# plot the smoothed timeseries with intervals
plt.figure(figsize=(16, 6))
plt.subplot(1,2,1)
plt.plot(df["date"], df["beta"], linewidth=3, color="blue", label="beta")
plt.title(f"Time Series for Beta", fontsize=20)
plt.legend(prop = {"size": 16 }, loc ="upper left")
plt.xlabel('Time', fontsize=14)
plt.ylabel('Transmission Rate', fontsize=14)
plt.xticks(ticks=df["date"][::15], rotation=45, fontsize=12)
plt.yticks(fontsize=12)


plt.subplot(1,2,2)
plt.plot(df["date"], df["nu"], linewidth=3, color="green", label="nu")
plt.title(f"Time Series for Nu", fontsize=20)
plt.legend(prop = {"size": 16 }, loc ="upper left")
plt.xlabel('Time', fontsize=14)
plt.ylabel('Vaccination Rate', fontsize=14)
plt.xticks(ticks=df["date"][::15], rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 6))

# First subplot: Case and PredCase
plt.subplot(1, 2, 1)
plt.scatter(df["date"], df["Case"], color="blue", label="Case")
plt.plot(df["date"], df["PredCase"], linewidth=3, color="red", label="PredCase")
plt.title(f"Time Series for Case", fontsize=20)
plt.legend(prop={"size": 16}, loc="upper left")
plt.xlabel('Time', fontsize=14)
plt.ylabel('Cases', fontsize=14)
plt.xticks(ticks=df["date"][::15], rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Second subplot: Vaccine and PredVaccine
plt.subplot(1, 2, 2)
plt.scatter(df["date"], df["Vaccine"],  color="green", label="Vaccine")
plt.plot(df["date"], df["PredVaccine"], linewidth=3, color="orange", label="PredVaccine")
plt.title(f"Time Series for Vaccine", fontsize=20)
plt.legend(prop={"size": 16}, loc="upper left")
plt.xlabel('Time', fontsize=14)
plt.ylabel('Vaccinations', fontsize=14)
plt.xticks(ticks=df["date"][::15], rotation=45, fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```
