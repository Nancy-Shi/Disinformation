##Visualization of Parameter Approximation Result by DNN##

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive',force_remount=True)
```

```python
df = pd.read_csv('/content/drive/My Drive/Misinformation_Project/neuralepi12_second.csv')
```

```python
# import libraries
import matplotlib.dates as mdates

# plot the smoothed timeseries with intervals
plt.figure(figsize=(16, 6))
plt.subplot(1,2,1)
df_omitted = df.iloc[:-3]
plt.plot(df_omitted ["date"], df_omitted ["beta"], linewidth=3, color="blue", label=r'$\beta$')
plt.title(r'Time Series for Transmission Rate ($\beta$)', fontsize=20)
#plt.legend(prop = {"size": 16 }, loc ="upper left")
plt.xlabel('Time', fontsize=18)
plt.ylabel(r'Transmission Rate ($\beta$)', fontsize=18)
plt.xticks(ticks=df_omitted ["date"][::15], rotation=45, fontsize=16)
plt.yticks(fontsize=16)


plt.subplot(1,2,2)
plt.plot(df_omitted ["date"], df_omitted ["nu"], linewidth=3, color="purple", label=r'$\nu$')
plt.title(r'Time Series for Vaccination Rate ($\nu$)', fontsize=20)
#plt.legend(prop = {"size": 16 }, loc ="upper left")
plt.xlabel('Time', fontsize=18)
plt.ylabel(r'Vaccination Rate ($\nu$)', fontsize=18)
plt.xticks(ticks=df_omitted ["date"][::15], rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout()
plt.show()
```

```python
df2 = pd.read_csv('/content/drive/My Drive/Misinformation_Project/neuralintervention_second.csv')
df3 = pd.read_csv('/content/drive/My Drive/Misinformation_Project/neuralvaccine_second_new.csv')
```

```python
plt.figure(figsize=(16, 7.5))
df2_omitted = df2.iloc[:-3]
df3_omitted = df3.iloc[:-3]


# First subplot: Intervention Information
plt.subplot(1, 2, 1)
plt.scatter(df2_omitted["date"], df2_omitted["MScoreInter"], edgecolors="blue", marker='o', label=r'Actual $M_1$', facecolors='none')
plt.plot(df2_omitted["date"], df2_omitted["PredMScoreInter"], linewidth=2, linestyle = "-.", color="blue", label=r'Predicted $M_1$')
plt.scatter(df2_omitted["date"], df2_omitted["CScoreInter"],  edgecolors="red", marker='o', label=r'Actual $C_1$', facecolors='none')
plt.plot(df2_omitted["date"], df2_omitted["PredCScoreInter"], linewidth=2,  linestyle = "--",color="red", label=r'Predicted $C_1$')
plt.title(f"Time Series for Intervention \n Information Impact Score", fontsize=20)
plt.legend(prop={"size": 16}, loc="upper left")
plt.xlabel('Time', fontsize=18)
plt.ylabel('Information Impact Score', fontsize=18)
plt.xticks(ticks=df2_omitted["date"][::15], rotation=45, fontsize=16)
plt.yticks(fontsize=16)


# First subplot: Vaccine Information
plt.subplot(1, 2, 2)
plt.scatter(df3_omitted["date"], df3_omitted["MScoreVac"], edgecolors="orange", marker='o', label=r'Actual $M_2$', facecolors='none')
plt.plot(df3_omitted["date"], df3_omitted["PredMScoreVac"], linewidth=2,linestyle = "-.", color="orange", label=r'Predicted $M_2$')
plt.scatter(df3_omitted["date"], df3_omitted["CScoreVac"],  edgecolors="purple", marker='o', label=r'Actual $C_2$', facecolors='none')
plt.plot(df3_omitted["date"], df3_omitted["PredCScoreVac"], linewidth=2, linestyle = "--", color="purple", label=r'Predicted $C_2$')
plt.title(f"Time Series for Vaccination \n Information Impact Score", fontsize=20)
plt.legend(prop={"size": 16}, loc="center left")
plt.xlabel('Time', fontsize=18)
plt.ylabel('Information Impact Score', fontsize=18)
plt.xticks(ticks=df3_omitted["date"][::15], rotation=45, fontsize=16)
plt.yticks(fontsize=16)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 6))

# First subplot: Case and PredCase
plt.subplot(1, 2, 1)
plt.scatter(df_omitted ["date"], df_omitted ["Case"], edgecolors="red", marker='o',  facecolors='none', label="Actual Cases")
plt.plot(df_omitted ["date"], df_omitted ["PredCase"], linewidth=2, linestyle = "-.", color="red", label="Predicted Cases")
plt.title(f"Time Series for Cases", fontsize=20)
plt.legend(prop={"size": 16}, loc="upper left")
plt.xlabel('Time', fontsize=18)
plt.ylabel('Total Cases', fontsize=18)
plt.xticks(ticks=df_omitted ["date"][::15], rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.gca().get_yaxis().offsetText.set_fontsize(16)

# Second subplot: Vaccine and PredVaccine
plt.subplot(1, 2, 2)
plt.scatter(df_omitted ["date"], df_omitted ["Vaccine"],  edgecolors="orange", marker='o',  facecolors='none', label="Actual Vaccine Uptake")
plt.plot(df_omitted ["date"], df_omitted ["PredVaccine"], linewidth=2, linestyle = "--", color="orange", label="Predicted Vaccine Uptake")
plt.title(f"Time Series for Vaccination", fontsize=20)
plt.legend(prop={"size": 16}, loc="upper left")
plt.xlabel('Time', fontsize=18)
plt.ylabel('Total Vaccine Uptake', fontsize=18)
plt.xticks(ticks=df_omitted ["date"][::15], rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.gca().get_yaxis().offsetText.set_fontsize(16)

plt.tight_layout()
plt.show()
```
